```python
%matplotlib inline
import pandas as pd
import numpy as np
import os
import matplotlib
import folium
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Open monster csv file 
csv_path=os.path.join("..","Sources", "monster_com-job_sample.csv")

jobs_data = pd.read_csv(csv_path)
jobs_data.columns

jobs_data_df=pd.DataFrame(jobs_data)

# Convert all data in job_title and job_description to lowercase 
jobs_data_df.loc[:,'job_title']=jobs_data['job_title'].str.lower()
jobs_data_df.loc[:,'job_description']=jobs_data['job_title'].str.lower()


#jobs_data_df.loc[jobs_data_df['organization'].isnull(),'Not Classified'] 
jobs_data_df.head(15)
```

```python
#Search for a job location (state) within the location column and save it to a new column and save to the dataframe jobs_data_df

state_search= ('AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL',
                  'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 
                  'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 
                  'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY')

pat = '|'.join(r"\b{}\b".format(x) for x in state_search)
jobs_data_df['state'] = jobs_data_df['location'].str.extract('('+ pat + ')', expand=False)
jobs_data_df = jobs_data_df[jobs_data_df['state'].notna()]

csv_path=os.path.join("..","Cleaned Data", "jobs.csv")
jobs_data_df.to_csv(csv_path)
```

```python
#Search for Data Science jobs
csv_path=os.path.join("..", "Cleaned Data", "jobs.csv")
ds_jobs_data_df = pd.read_csv(csv_path)


job_title_search=("data sci", "machine learn", "data analy", "data min", "statist",
                 "big data", "business intelligence", "analytic","data engineer", "business analy",
                  "database admin", "data specia")

pat = '|'.join (job_title_search)

ds_jobs_data_df=jobs_data_df.loc[jobs_data_df['job_title'].str.contains(pat), :]

ds_jobs_data_df.head()
```

```python
#Clean up the Industry categories

ds_jobs_data_df.loc[ds_jobs_data_df['organization'].isnull(), 'organization']= ds_jobs_data_df['sector'] 
ds_jobs_data_df.loc[ds_jobs_data_df['organization']=='All', 'organization']= ds_jobs_data_df['sector'] 

ds_jobs_data_df.head()
```

```python
#Reclassify Industries

ds_jobs_data_df.loc[ds_jobs_data_df.organization == 'Biotech/R&D/Science', 'organization']='Biotechnology/Pharmaceuticals' 
ds_jobs_data_df.loc[ds_jobs_data_df.organization =='Computer/IT Services','organization']='Technology' 
ds_jobs_data_df.loc[ds_jobs_data_df.organization =='IT/Software Development','organization']='Technology' 
ds_jobs_data_df.loc[ds_jobs_data_df.organization =='Computer Hardware','organization']='Technology' 
ds_jobs_data_df.loc[ds_jobs_data_df.organization =='Experienced (Non-Manager)','organization']='Other/Not Classified' 

csv_path=os.path.join("..","Cleaned Data", "ds_jobs.csv")
ds_jobs_data_df.to_csv(csv_path)
```
